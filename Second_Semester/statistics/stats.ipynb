Проанализируем данные с эксперимента по измерению массы номинально одинаковых грузиков.

```python
import pandas
import matplotlib.pyplot as plt
from scipy import stats


masses = pandas.read_csv('datasets/masses.csv') 
masses['experiments'].plot(kind='bar')

df = pandas.DataFrame(data={
    'Mass': masses['experiments'],
})

df.plot.kde()

plt.show()

d = df['Mass']

print(stats.kstest(d, 'norm', (d.mean(), d.std()), N=5000)[0])
```
